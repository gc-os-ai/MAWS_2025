### Surface-aware sampling parameters (new in 2026)

MAWS now restricts the initial pose search to the region just outside the ligand surface, computed from the ligand atoms automatically. Three knobs:

- **`shape`** — geometry of the search envelope around the ligand. `shell` (default) hugs the surface most tightly; `sphere` is a simple bounding sphere; `cube` is the bounding box.
- **`reach`** — how far past the ligand surface the envelope extends, in ångströms. Roughly the distance an aptamer atom can sit beyond the protein. Default `10.0` Å — about one nucleotide's worth of slack.
- **`probe`** — radius of the rolling probe used to define the SAS surface, in ångströms. A candidate pose is rejected if any aptamer atom would land within `vdW(atom) + probe` of any protein atom. Default `1.4` Å (water-equivalent, the convention used by Chimera/PyMOL/FreeSASA).

Keep the defaults unless your target is unusual (very small ligand, very deep pocket, etc.).

```python
# Job name
job_name = "1brq_maws_aptamer_final"

# Number of nucleotides in the aptamer (default: 15)
nucleotides = 3

# Path to your pdb file
my_path = "/home/siddharth899/Work/ESOC/MAWS_2025/data/1BRQ.pdb"

# Aptamer type (DNA or RNA)
apt_type = "DNA"

# Molecule type (protein, organic or lipid)
mol_type = "protein"

# Name of your conda environment
env_name = "maws"

# Beta (default: 0.01)
beta = 0.01

# First chunk size (default: 5000)
chunk_1 = 5000

# Second chunk size (default: 5000)
chunk_2 = 5000

# Sampling envelope shape: cube | sphere | shell (default: shell)
shape = "shell"

# How far the envelope extends beyond the ligand surface, in Å (default: 10)
reach = 10.0

# vdW probe radius for SAS rejection, in Å (default: 1.4, water-like)
probe = 1.4
```

```python
!python ../../maws/maws2023.py \
  --name "{job_name}" \
  --ntides {nucleotides} \
  --path "{my_path}" \
  --aptamertype "{apt_type}" \
  --moleculetype "{mol_type}" \
  --beta {beta} \
  --firstchunksize {chunk_1} \
  --secondchunksize {chunk_2} \
  --clean-pdb \
  --remove-h \
  --drop-hetatm \
  --shape {shape} \
  --reach {reach} \
  --probe {probe}
```
